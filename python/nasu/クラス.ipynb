```python
class Person:
  def __init__(self,name,nationality,age):
    self.name = name
    self.nationality = nationality
    self.age = age

  def  __call__(self):
    print('ここはcall関数です。')

  def say_hello(self,name):
    print('こんにちは、{}さん。私は{}です。'.format(name,self.name))
```

```python
#インスタンス化
nasu = Person('那須','日本',27)
```

```python
nasu()
```
