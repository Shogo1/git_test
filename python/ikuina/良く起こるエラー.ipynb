SyntaxError
構文エラー

NameError

定義していない変数名を使用した際に起こるエラー

AttributeError

オブジェクトが持ってない属性を使用した際に起こるエラー

TypeError

謝ったデータ型同士の計算等で起こるエラー

ValueError

謝った値を取っている場合に起こるエラー

IndentationError

インデント周りで起こるエラー、
インデントがズレていたり誤ってると起こるエラー

IndexError

範囲外のインデックス（要素番号）を指定した際に起こるエラー

KeyError

辞書で登録されていないキーを指定した際に起こるエラー

ModuleNotFoundError

importで指定したモジュールがないエラー

FileNotFoundError

指定したファイルが存在しないエラー
