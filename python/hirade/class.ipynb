```python
class Person:

  def __init__(self,name, nationality, age):
    self.name = name
    self.nationality = nationality
    self.age = age

  def __call__(self):
    print('ここはcall関数です。')

  def say_hello(self,name):
    print('こんにちは。{}さん。私は{}です。'.format(name,self.name))
```

```python
#インスタンス化(実体化)
imanishi = Person('今西','日本', 25)
```

```python
imanishi.name
```

```python
imanishi.say_hello('佐藤')
```

```python
imanishi()
```
