```python
last_names = ['今西','佐藤','鈴木','田中']
first_names = ['航平','謙介', '康二','康介']
```

```python
for i in range(len(last_names)):
  print(last_names[i] + first_names[i]+ 'さん')
```

```python
for last_name, first_name in zip(last_names, first_names):
  print(last_name + first_name + 'さん')
```
